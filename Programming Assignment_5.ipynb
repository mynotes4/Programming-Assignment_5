```python
# 1.Write a Python Program to Find LCM?
a = int(input("Enter First Number "))
b = int(input("Enter Second Number "))
f = 1
m = max([a,b])
l=[]
for i in range(2,m+1):
    for j in range(2,int(i/2)):
        if i%j == 0:
            break
    else:
        l.append(i)
while a > 1 or b> 1:
    for i in l:
        if a%i == 0 or b%i == 0:
            f = f*i
        if a%i == 0:
            a = a/i
        if b%i == 0:
            b = b/i

print("LCM ",f)
```

```python
#2. Write a Python Program to Find HCF?
a = int(input("Enter First Number "))
b = int(input("Enter Second Number "))
a1 = a
b1 = b
f = 1
m = max([a,b])
l=[]
for i in range(2,m+1):
    for j in range(2,int(i/2)):
        if i%j == 0:
            break
    else:
        l.append(i)
while a > 1 or b> 1:
    for i in l:
        if a%i == 0 or b%i == 0:
            f = f*i
        if a%i == 0:
            a = a/i
        if b%i == 0:
            b = b/i
hcf = a1*b1/f
print("hcf",hcf)
```
